```python
#On compte le nombre de mails qu'on analyse, ainsi que le nombre de spam/ham.

ham=0
spam=0
n=0
label=[]
for x in D_train["label"]:

    if x =='ham':
        ham+=1
        label.append(1)
    if x=='spam':
        spam+=1
        label.append(0)
 
    n+=1

        
        
print("ham = ", ham," / ","spam = ",spam," donc ",ham+spam," sur ",n,)

#Calcul de la probablité P(Ham) et P(Spam) de recevoir un ham/spam : 

P_spam = spam/(ham+spam)
P_ham = ham/(ham+spam)

P_spam, P_ham

```

```python

nb=0
spamMot={} #Dictionnaire des mots contenus dans les mails marqués spam
hamMot={}  #Dictionnaire des mots contenus dans les mails marqués ham
dictmot={} #Dictionnaire de tous les mots vus ainsi que de leurs nombres d'occurences. 
mmoott=set()
nbrHam=0
nbrSpam=0

n=0
for x in D_train["text"]:
    ligne=[]
    x=x[9:]
    mot=""
    for y in x:
        if y==' ' or y=='\r':
            if(len(mot)>4):
                if(label[nb]==0): #Si c'est un mot qui est dans un Ham
                    try:
                        hamMot[mot]+=1
                    except:
                        hamMot[mot]=1
                    
                else: #Sinon le mail initial est un spam 
                    try:
                        spamMot[mot]+=1
                    except:
                        spamMot[mot]=1
                    
                try: #On le rajoute dans le dictionnaire des mots -qui réuni tous les mots analysés dans les mails 
                    #et on comptera les occurences de ceux-ci lors du parcours des données
                    
                    a,b=dictmot[mot]
                    if(label[nb]==0):
                        b+=1
                    else:
                        a+=1
                    dictmot[mot]=(a,b)
                        
                except:
                    if(label[nb]==0):
                        dictmot[mot]=(0,1)
                    else:
                        dictmot[mot]=(1,0)
            if(label[nb]==0):
                nbrHam+=1
            else:
                nbrSpam+=1
            mmoott.add(mot)
            mot=""
            n+=1
        else:
            if y!='\n':
                mot+=y
    nb+=1
    
    
print(nbrSpam)#Nombre de mots totaux lus dans les Spam
print(nbrHam) #Nombre de mots totaux lus dans les Ham  
print(n) #Nombre de mots totaux lus (comprend des doublons)
print(len(dictmot)) #Nombre de mots au total dans le dictionnaire
nbrMots=len(mmoott) 
```

```python
#Tout au long de nos analyses, on a calculé les facteurs et probabilité suivante 

#P_spam, P_ham - Probabilités de recevoir un spam ou ham 
#N_spam = nbrSpam - Le nombre de mots au total dans tous les mails de spam 
#N_ham = nbrHam -  Le nombre de mots au total dans tous les mails de ham 
#N_dico - Le nombre de mots dans le dictionnaire 

alpha = 1 #Qui est le facteur pour la formule de laplace 


#Calcul pour chaque mot de P(mot | Spam) et P(mot | ham), qui seront utilisés dans la formule de Bayes pour 
#P(Spam| plusieurs mots ) .

#Choix du vecteurs sur tous les mots ici : 
#C'est énorme, mais à revoir. 
p_mot_spam= {mot : 0 for mot in dictmot}
p_mot_ham = {mot : 0 for mot in dictmot}

for mot in dictmot :
    x,y=dictmot[mot]
    #On parcours le dictionnaire, et on calcule les probabilités d'avoir ce mot en sachant que notre mail est un spam ou ham. 
    p_mot_ham[mot]=(y+alpha)/(nbrHam + alpha* n) 
    p_mot_spam[mot]=(x+alpha)/(nbrSpam + alpha* n)
    
def bernoulli(X,p):
    #Calcule P(X=x) où X qui suit une loi de Bernoulli de paramètre p
    return (p**X)*(1-p)**(1-X)


## Fonction "brouillon" rajouté trop tardivement pour adapter le code où on calcule la somme des logs de p(x|y)... : 

def log_P(X,d): 
    p=[]
    for i in range(d.size): 
        p.append(bernouille(X[:,i],d[i]))
    p=np.array(p).transpose()
    log_p=np.log(p)
    
    return np.sum(log_p,axis=1) #La somme des logs de p(x|y) selon le taux de xi
```

### Création de la fonction de notre modèle : 

```python

def antispam(mail):
    #on met en minuscule le message du mail pour faciliter la lecture (car nos mots dans le dictionnaire était en minuscule) 
    #puis on split ce message pour récuperer chacun des mots et faire le traitement avec la formule de bayes    
    mail = mail.lower().split()

    #Probabilité que le message donné soit un spam ou un ham, on pose au départ ces probabilités d'après la formule.
    
    
    P_ham_mail= P_ham 
    P_spam_mail = P_spam

    for mot in mail:
        x=mot in p_mot_spam
        
        '''
        Si on ne retrouve pas le mot, on ne met pas le facteur à 0 sinon cela fausse le résultat  
        Une application du logarithme aurait pu se faire notamment ici pour simplifier... 
        '''
        
        p=bernoulli(x,p_mot_spam[mot] if mot in p_mot_spam else 0.0001) 
        P_spam_mail *= p
        #print(mot)
        
        x=mot in p_mot_ham
        p=bernoulli(x,p_mot_ham[mot] if mot in p_mot_ham else 0.0001)
        P_ham_mail *= p
        #print(mot)

    #print(p_ham_mail, p_spam_mail) 

    if P_ham_mail > P_spam_mail : 
        #print("Ce message est suscpetible d'être un ham à ", P_ham_mail,"%")
        return 1
    elif P_ham_mail <= P_spam_mail : 
        #print("Ce message est suscpetible d'être un spam à ", P_spam_mail,"%")
        return 0
    else : #Probabilité égale d'être un mail ou spam, dans le pire des cas.. 
        print("Il y a autant de chance d'être un spam ou un mail, veillez vérifier avec précaution.")
        return -1
    
```
